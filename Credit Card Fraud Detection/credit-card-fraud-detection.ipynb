```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from xgboost import XGBClassifier
from sklearn.linear_model import SGDClassifier
from sklearn import svm
import lightgbm as ltb
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("../input/creditcardfraud/creditcard.csv")
df
```

<font size="5">correlation matrix of data</font>

```python
plt.rcParams["figure.figsize"] = (22, 12)
corr=df.corr()
sns.heatmap(corr,annot=True)
```

<font size="5">split data</font>

```python
features=['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14','V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23','Class', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount',
       'Class']
X=df.drop(["Class"],axis=1)
Y=df["Class"]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=42)
```

<font size="5">Models used for detection :</font>
* 1-KNeighborsClassifier
* 2-XGBoost classifier
* 3-SGDClassifier
* 4-LGBMClassifier
* 5-RandomForestClassifier




```python
KN=KNeighborsClassifier(n_neighbors=2)
KN.fit(X_train,Y_train)
K_predict=KN.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, K_predict, labels=KN.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=KN.classes_)
plt.rcParams["figure.figsize"] = (12, 6)
disp.plot(cmap=plt.cm.Blues)
print("f1_score : "+str(f1_score(Y_test,K_predict)))
print("roc_auc_score : "+str(roc_auc_score(Y_test, K_predict)))
```

```python
xgb=XGBClassifier()
xgb.fit(X_train,Y_train)
xgb_pred=xgb.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, xgb_pred, labels=xgb.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=xgb.classes_)
plt.rcParams["figure.figsize"] = (12, 6)
disp.plot(cmap=plt.cm.Blues)
print("f1_score : "+str(f1_score(Y_test, xgb_pred)))
print("roc_auc_score : "+str(roc_auc_score(Y_test, xgb_pred)))
```

```python
SGD=SGDClassifier(penalty=None)
SGD.fit(X_train,Y_train)
SGD_pred=SGD.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, SGD_pred, labels=SGD.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=SGD.classes_)
plt.rcParams["figure.figsize"] = (12, 6)
disp.plot(cmap=plt.cm.Blues)
print("f1_score : "+str(f1_score(Y_test, SGD_pred)))
print("roc_auc_score : "+str(roc_auc_score(Y_test, SGD_pred)))
```

```python
L_gbm = ltb.LGBMClassifier()
L_gbm.fit(X_train, Y_train)
L_gbm_pred = L_gbm.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, L_gbm_pred, labels=L_gbm.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=L_gbm.classes_)
plt.rcParams["figure.figsize"] = (12, 6)
disp.plot(cmap=plt.cm.Blues)
print("f1_score : "+str(f1_score(Y_test, L_gbm_pred)))
print("roc_auc_score : "+str(roc_auc_score(Y_test, L_gbm_pred)))
```

```python
RFC = RandomForestClassifier(max_depth=10, random_state=0)
RFC.fit(X_train, Y_train)
RFC_pred = RFC.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, RFC_pred, labels=RFC.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=RFC.classes_)
plt.rcParams["figure.figsize"] = (12, 6)
disp.plot(cmap=plt.cm.Blues)
print("f1_score : "+str(f1_score(Y_test, RFC_pred)))
print("roc_auc_score : "+str(roc_auc_score(Y_test, RFC_pred)))
```
